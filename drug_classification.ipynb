```python
from __future__ import absolute_import, division, print_function, unicode_literals
import tensorflow as tf
from google.colab import files
import io
import pandas as pd
```

```python
uploaded = files.upload()
```

```python
train_data = pd.read_csv(io.BytesIO(uploaded['drug200.csv']))
```

```python
train_data.head()
```

```python
train_data.head()
```

```python
train_y = train_data.pop('Drug')
```

```python
id = train_y.unique()
```

```python
feature_columns = []
categorical_columns = ['Sex', 'BP', 'Cholesterol']
numeric_columns = ['Age', 'Na_to_K']
for feature in categorical_columns:
  keys = train_data[feature].unique()
  feature_columns.append(tf.feature_column.embedding_column(tf.feature_column.categorical_column_with_vocabulary_list(feature, keys), dimension = 8))
for feature in numeric_columns:
  feature_columns.append(tf.feature_column.numeric_column(feature, dtype = tf.float32))
```

```python
def encode(label):
  keys = label.unique()
  index = 0
  index_dict = {}
  for key in keys:
    index_dict[key] = index
    index += 1
  for i in range(label.shape[0]):
    label[i] = index_dict[label[i]]
  return label
```

```python
train_y = encode(train_y)
```

```python
train_y = pd.to_numeric(train_y)
```

```python
train_y.head()
```

```python
def input_fn(features, labels, shuffle = True, batch_size = 256):
  dataset = tf.data.Dataset.from_tensor_slices((dict(features), labels))
  if shuffle:
    dataset = dataset.shuffle(1000).repeat()
  return dataset.batch(batch_size)
```

```python
classifier = tf.estimator.DNNClassifier(feature_columns = feature_columns, hidden_units = [50, 15], n_classes = 5)
```

```python
classifier.train(input_fn = lambda: input_fn(train_data, train_y, shuffle = 1), steps = 15000)
```

```python
result = classifier.evaluate(input_fn = lambda: input_fn(train_data, train_y,shuffle = False))

print(result)
```
